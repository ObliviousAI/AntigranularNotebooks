## Distribution of Columns üìä

```python
df_hist = df.hist(figsize = (10, 10), rwidth = 0.95, color = "skyblue", grid = False)
plt.title("Distributions")
```

The above was to see the distribution from a regular DataFrame, but from a PrivateDataFrame, the above code will not work. This is because of the following reasons: ü¶æ
* We do not support **matplotlib** library in the AG kernel.
* We do not support plotting of any graphs in the AG kernel.
* The *hist* method only works on 1 column at a time for a PrivateDataFrame.

So how can you see the distribution of a PrivateDataFrame? ü§® We will have to first execute the *hist* command on each of the column. It will return the frequency and bins for that column. ü§ùüèº

```python
%%ag
hist_data = {}
hist_data["Age"] = pdf.hist(figsize = (10, 10), rwidth = 0.95, color = "skyblue", grid = False, column="Age", eps=0.01, bins=10)
hist_data["Education Level"] = pdf.hist(figsize = (10, 10), rwidth = 0.95, color = "skyblue", grid = False, column="Education Level", eps=0.01, bins=4)
hist_data["Years of Experience"] = pdf.hist(figsize = (10, 10), rwidth = 0.95, color = "skyblue", grid = False, column="Years of Experience", eps=0.01, bins=10)
hist_data["Salary"] = pdf.hist(figsize = (10, 10), rwidth = 0.95, color = "skyblue", grid = False, column="Salary", eps=0.01, bins=10)
hist_data["Senior"] = pdf.hist(figsize = (10, 10), rwidth = 0.95, color = "skyblue", grid = False, column="Senior", eps=0.01, bins=2)
```

We will now export this data back to our local kernel, where we can plot the data received from the *hist* method executed above. ü•≥

By exporting, we are just copying the variable from the Antigranular environment to our local environment.

Note, the use of epsilon for the *hist* method. It will return differentially private result for the *hist* method. The epsilon value, dictates on how much noise will be added to the result to make it differentially private. üîê

Epsilon is an important metrics which measures your privacy budget. More the budget you spend, the better utility of the data but also a higher risk of compromised privacy. ü§ì

Basically, the lower the epsilon value, the more private your values are. This is a delicate privacy vs. utility trade-off you will learn to navigate when working with differential privacy. üö®

```python
%%ag
export(hist_data,"hist_data")
```

```python
print(hist_data)
```

We will perform the same action on the DataFrame in our local kernel using the method **histogram** of library *numpy*. We do this to see the data in form of graphs which are similar to what PrivateDataFrame returns. üìÄ The numpy historgram, provides us with the data in as similar way, which will make the visual comparison easier.

```python
df_hist_data = {}
df_hist_data["Age"] = np.histogram(df['Age'], bins=10)
df_hist_data["Education Level"] = np.histogram(df['Education Level'], bins=4)
df_hist_data["Years of Experience"] = np.histogram(df['Years of Experience'], bins=10)
df_hist_data["Salary"] = np.histogram(df['Salary'], bins=10)
df_hist_data["Senior"] = np.histogram(df['Senior'], bins=2)
```

We will now plot both the DataFrame and PrivateDataFrame, side by side, to see the difference between the results. ü§© You will also learn how to plot the graph from the results of *hist* method of PrivateDataFrame.

```python
#Seperating frequencies and edges of PrivateDataFrame result
hist_df = pd.DataFrame(hist_data)
pdf_frequencies = [hist_df[column][0] for column in hist_df.columns]
pdf_bin_edges  = [hist_df[column][1] for column in hist_df.columns]

#Seperating frequencies and edges of DataFrame result
df_frequencies = [df_hist_data[col][0] for col in df_hist_data]
df_bin_edges  = [df_hist_data[col][1] for col in df_hist_data]

#We are plotting for 5 columns, a histogram for each column in a seperate row
n_rows = 5
fig, axes = plt.subplots(n_rows, 2, figsize=(10, n_rows*4),) #setting 2 columns, one for DataFrame and one for PrivateDataFrame

#Adjusting space between the graphs
plt.subplots_adjust(wspace=10, hspace=10)

for i,col in enumerate(hist_df.columns):
    # Calculate bin widths and centers
    pdf_bin_widths = np.diff(pdf_bin_edges[i])
    pdf_bin_centers = pdf_bin_edges[i][:-1] + pdf_bin_widths / 2
    df_bin_widths = np.diff(df_bin_edges[i])
    df_bin_centers = df_bin_edges[i][:-1] + df_bin_widths / 2

    # Plotting for DataFrame
    axes[i][0].bar(df_bin_centers, df_frequencies[i], width=df_bin_widths, align='center', color = "skyblue", edgecolor="black")
    axes[i][0].set_xlabel('Value')
    axes[i][0].set_ylabel('Frequency')
    axes[i][0].set_title(f'Histogram for \"{col}\"')

    # Plotting for PrivateDataFrame
    axes[i][1].bar(pdf_bin_centers, pdf_frequencies[i], width=pdf_bin_widths, align='center', color = "red", edgecolor="black",)
    axes[i][1].set_title(f'Histogram for Private \"{col}\"')
    axes[i][1].set_xlabel('Value')
    axes[i][1].set_ylabel('Frequency')


# Add borrder to the plots
for ax in axes.flat:
    for _, spine in ax.spines.items():
        spine.set_visible(True)
        spine.set_linewidth(2)
        spine.set_edgecolor('black')

# Add background
fig.patch.set_facecolor('lightgrey')

plt.tight_layout()
plt.show()
```

See the difference between the graphs with data from DataFrame and PrivateDataFrame? Because of the epsilon, the distribution for different columns varies. üßëüèº‚Äçüíª

For some columns, like **Senior** and **Education Level**, the floating point numbers don't make sense. But in the Antigranular kernel, the values generated have noise added in them üì¢, thus there are some data points which have float values, instead of the expected int values in the provided range.

In Antigranular, for a competion or a dataset, in the description section, the unique values for all the **categorical** data are provided. ü§ì But exposing the unique values for numerical data, defeats the purpose of what Antigranular is all about, which is protecting the sensitive information.

A few more methods are **restricted** on a PrivateDataFrame on Antigranular.
‚ùå They are:

* **min** - Minimum ( Work around quantile(0) )

* **max** - Maximum ( Work around quantile(1.0) )

* **mode** - Mode

* **abs** - Absolute Value

* **prod** - Product of values

* **sem** - Standard error of the mean

* **kurt** - Sample kurtosis (4th moment)

* **cumsum** - Cumulative sum

* **cumprod** - Cumulative product

* **cummax** - Cumulative maximum

* **cummin** - Cumulative minimum



The ones **allowed**, all require an epsilon value passed to them. The usage of these methods can be found [here](https://docs.antigranular.com/private-python/api/op_pandas/private_dataframe). üìñ

## Correlation Matrix üå™Ô∏è

Correlation method gives the information about how two columns might be related to each other.

The difference between the correlation method of DataFrame and PrivateDataFrame is only the usage of epsilon. Take a look at how the result differs with different epsilon values. üî¨

```python
df_correlation_matrix = df.corr()
correlation_value = df_correlation_matrix.loc['Age', 'Salary']

print("Correlation Matrix:\n", df_correlation_matrix)
print("\nCorrelation between Age and Salary:", correlation_value)
```

```python
%%ag
pdf_correlation_matrix_0_01 = pdf.corr(eps = 0.01)
correlation_value = pdf_correlation_matrix_0_01.loc['Age', 'Salary']

ag_print("Correlation Matrix:\n", pdf_correlation_matrix_0_01)
ag_print("\nCorrelation between Age and Salary:", correlation_value)
```

```python
%%ag
pdf_correlation_matrix_1 = pdf.corr(eps = 1)
correlation_value = pdf_correlation_matrix_1.loc['Age', 'Salary']

ag_print("Correlation Matrix:\n", pdf_correlation_matrix_1)
ag_print("\nCorrelation between Age and Salary:", correlation_value)
```

For better visualisation, let's view all these matrices together as heatmaps. üî• For that we will export the matrixes inside the Antigranular kernel to local kernel. ‚¨áÔ∏è

```python
%%ag
export(pdf_correlation_matrix_1,"pdf_correlation_matrix_1")
export(pdf_correlation_matrix_0_01, "pdf_correlation_matrix_0_01")
```

We are removing NaN values as validating the type of Correlation results. For PrivateDateFrame, when calculating correlation with higher noise, you might get NaN values. ‚≠ïÔ∏è To plot the graphs, we need to replace these with some valid values, so replacing it with a 0 here. Also to make sure, we don't values of any other type, we convert the result to *float* type.

```python
pdf_correlation_matrix_0_01 = pdf_correlation_matrix_0_01.fillna(0).astype(float)
pdf_correlation_matrix_1 = pdf_correlation_matrix_1.astype(float)
```

Now, let's rename columns and rows for better chart visualisation.

Notice that in correlation matrix, both the rows and column names represent the actual column names from the dataframe. üëçüèΩ

```python
df_correlation_matrix = df_correlation_matrix.rename(columns={"Education Level":"EL","Years of Experience": "YE"}, index={"Education Level":"EL","Years of Experience": "YE"})
pdf_correlation_matrix_0_01 = pdf_correlation_matrix_0_01.rename(columns={"Education Level":"EL","Years of Experience": "YE"}, index={"Education Level":"EL","Years of Experience": "YE"})
pdf_correlation_matrix_1 = pdf_correlation_matrix_1.rename(columns={"Education Level":"EL","Years of Experience": "YE"}, index={"Education Level":"EL","Years of Experience": "YE"})
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15 , 4))
sns.heatmap(df_correlation_matrix, ax=axes[0], annot=True)
sns.heatmap(pdf_correlation_matrix_0_01, ax=axes[1], annot=True)
sns.heatmap(pdf_correlation_matrix_1, ax=axes[2], annot=True)

axes[0].set_title("DataFrame Corr Heatmap")
axes[1].set_title("PrivateDataFrame Corr Heatmap eps=0.01")
axes[2].set_title("PrivateDataFrame Corr Heatmap eps=1")

# Add borrder to the plots
for ax in axes.flat:
    for _, spine in ax.spines.items():
        spine.set_visible(True)
        spine.set_linewidth(2)
        spine.set_edgecolor('black')

# Add background
fig.patch.set_facecolor('lightgrey')

plt.tight_layout()
plt.show()
```

From putting the heatmaps side by side üëÄ, we can see that the results we received with an epsilon value of 0.01 are far away from the actual results. üòÖ On the other hand, the heatmap with result of epsilon value 1 are way closer to the actual results. üíÉüèº

The more epsilon you spend, the better results you get. üôèüèΩ But for the competitions, one must balance the epsilon usage and the amount of information you can gather. ü§® You can also refer to the notebook [**Understanding Performance Over Epsilon Values**](https://www.antigranular.com/notebooks/654778023bb10c5b07d6b8ee) to understand this concept better.
